```python
df2['extras_type'] = df2['extras_type'].fillna('No Extra')
df2['player_dismissed'] = df2['player_dismissed'].fillna('Not Dismissed')
df2['dismissal_kind'] = df2['dismissal_kind'].fillna('Not Dismissed')
df2['fielder'] = df2['fielder'].fillna('No Fielder')
```

```python
df2.isnull().sum()
```

```python
import matplotlib.pyplot as plt

top_batsmen = df2.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)

top_batsmen.plot(kind='bar', color='red')
plt.title('Top 10 Batsmen by Runs')
plt.ylabel('Runs')
plt.xlabel('Batter')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filter only rows where a player was dismissed
wickets = df2[df2['player_dismissed'] != 'Not Dismissed']

# Count wickets per bowler
top_bowlers = wickets['bowler'].value_counts().head(10)
top_bowlers.plot(kind='bar', color='green')
plt.title('Top 10 Wicket-Takers')
plt.ylabel('Wickets')
plt.xlabel('Bowler')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
